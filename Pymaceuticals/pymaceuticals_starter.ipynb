```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
import matplotlib.ticker as mticker

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_data, clinical_trial_data, on='Mouse ID', how = 'outer')

# Display the data table for preview

combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume_mean = combined_data.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)'].reset_index()
tumor_volume_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_errors = combined_data.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)'].reset_index()
standard_errors = standard_errors.rename(columns={'Tumor Volume (mm3)':'Error'})

standard_errors.head()
```

```python
standard_errors_p = standard_errors.pivot(columns= 'Drug',index= 'Timepoint', values='Error')
```

```python
# Minor Data Munging to Re-Format the Data Frames
data_munging = tumor_volume_mean.pivot(columns= 'Drug',index= 'Timepoint', values='Tumor Volume (mm3)')

data_munging.head()
```

```python
# Generate the Plot (with Error Bars)
data_munging_plot = data_munging[['Capomulin','Infubinol','Ketapril','Placebo']]

fig, ax = plt.subplots()
data_munging_plot.plot(ax=ax,yerr=standard_errors_p, ecolor='black',elinewidth=.5,linestyle=':',linewidth=.8,)

Capomulin, = plt.plot(data_munging['Capomulin'], marker="o",color="red", linestyle=':',linewidth=.8,  label="Capomulin")
Infubinol, = plt.plot(data_munging['Infubinol'], marker="^",color="blue", linestyle=':',linewidth=.8, label="Infubinol")
Ketapril, = plt.plot(data_munging['Ketapril'], marker="s",color="green", linestyle=':',linewidth=.8, label="Ketapril")
Placebo, = plt.plot(data_munging['Placebo'], marker="d",color="black", linestyle=':',linewidth=.8, label="Placebo")

y = [35,40,45,50,55,60,65,70]
plt.hlines(y, -2, 47, alpha=0.15)
plt.xlim(-2, 47)
plt.ylim(32, 74)
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.legend(handles=[Capomulin,Infubinol,Ketapril,Placebo], loc='best')
# Save the Figure
plt.savefig("Tumor Response to Treatment")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
final_f = data_munging[['Capomulin','Infubinol','Ketapril','Placebo']]

# change_df = pd.DataFrame({'Drug':name['Drug'],'Percent Change':change})
final_g = (final_f.unstack().groupby("Drug").last()-final_f.unstack().groupby("Drug").first())/45*100
final_g
```

```python
list1 = final_g.values
# myList = [round(x) for x in list1]
myList = ['{:.0f}%'.format(x) for x in list1]
```

```python
# Store all Relevant Percent Changes into a Tuple
fig, ax = plt.subplots(figsize=(8, 6))
my_colors = ['g','r','r','r']
ax = final_g.plot(kind ='bar',width=1, color=my_colors, rot=0, position=1)

ax.text(-.6,-4,myList[0],color='w',fontsize=10)
ax.text(.4,3,myList[1],color='w',fontsize=10)
ax.text(1.4,3,myList[2],color='w',fontsize=10)
ax.text(2.4,3,myList[3],color='w',fontsize=10)

tick_spacing=20
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.set_ylabel('Tumor Volume (mm3)')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_ylim(-30,70)
ax.grid()
# Save the Figure
plt.savefig("Tumor Change Over 45 Day Treatment")
# Show the Figure
fig.show()
```
